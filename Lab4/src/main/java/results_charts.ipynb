```python
import pandas
import matplotlib.pyplot as plt

data = pandas.read_csv("../../../naiveOutput.csv",header=None)
producers = data.loc[data[0] == "Producer"]
consumers = data.loc[data[0] == "Consumer"]

```

```python
plt.figure()
plt.stem(producers[1].unique()[:1000],producers[2].groupby(producers[1]).mean()[:1000],label="Producer")
```

```python
plt.figure()
plt.stem(consumers[1].unique()[:1000],consumers[2].groupby(consumers[1]).mean()[:1000],label="Consumer")
```
